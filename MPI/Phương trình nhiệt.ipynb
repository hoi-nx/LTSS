```python
import matplotlib.pyplot as plt
import numpy as np
%pylab inline

x = np.arange(0, 50)
y = np.arange(0, 50,)
X, Y = np.meshgrid(x, y)


extent = np.min(x), np.max(x), np.min(y), np.max(y)
fig = plt.figure(frameon=False)

Z1 = np.add.outer(range(50), range(50)) % 2  # chessboard
im1 = plt.imshow(Z1, cmap=plt.cm.gray, interpolation='nearest',alpha=0.3,extent=extent)

Z2 = df.dropna(axis='columns').values
#cmap=hot
im2 = plt.imshow(Z2, cmap=plt.cm.Reds, alpha=.9, interpolation='bilinear',
                 extent=extent)
plt.colorbar() 

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
import numpy as np
from matplotlib.ticker import LinearLocator, FormatStrFormatter   

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection = '3d')
jet = plt.get_cmap('jet')

x = np.arange(0,50,1)
y = np.arange(0,50,1)
X, Y = np.meshgrid(x, y)

Z = np.array(df.dropna(axis='columns').values)

surf = ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = jet, linewidth = 1)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
ax.set_zlim3d(0, Z.max())

fig.colorbar(surf, shrink=0.4, aspect=4)





plt.show()
```

```python
A = df.dropna(axis='columns').values

fig, axs = plt.subplots(1, 3, figsize=(10, 3))
for ax, interp in zip(axs, ['nearest', 'bilinear', 'bicubic']):
    ax.imshow(A, interpolation=interp)
    ax.set_title(interp.capitalize())
    ax.grid(True)

plt.show()
```
